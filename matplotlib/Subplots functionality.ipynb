```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
a = np.linspace(0,10,11)
b = a ** 4
```

```python
x = np.arange(0,10)
y =x*x
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(a,b)
axes[0][1].plot(x,y)
axes[1][0].plot(b,a)
axes[1][1].plot(y,x)
fig.suptitle("Figure",fontsize=15)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(a,b)
axes[0][1].plot(x,y)
axes[1][0].plot(b,a)
axes[1][1].plot(y,x)
fig.suptitle("Figure",fontsize=15)
fig.subplots_adjust(wspace=0.5,hspace=0.5)
fig.savefig('subplots.png',bbox_inches='tight')
```
