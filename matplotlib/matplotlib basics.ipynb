```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x=np.arange(0,20)
```

```python
y=x*x
```

```python
plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xlim(0,9)
plt.ylim(0,100)
plt.title('Sample')
# to save as raw image
plt.savefig('My First plot graph.png')
```
