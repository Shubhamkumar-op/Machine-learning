```python
a=np.linspace(0,10,20)
```

```python
b=a**3
```

```python
x=np.arange(0,100,5)
y=2*x
```

```python
n2=plt.figure(figsize=(10,10))
#large axes
axes1=n2.add_axes([0,0,1,1])
axes1.plot(a,b)
axes1.set_xlabel("A----->")
axes1.set_ylabel("B----->")
axes1.set_title("FIGURE 1")
#small axes
axes2=n2.add_axes([0.1,0.1,0.5,0.5])
axes2.plot(x,y);
axes2.set_xlabel("X----->")
axes2.set_ylabel("Y----->")
axes2.set_title("FIGURE 2")
n2.savefig('MULTIGRAPH',bbox_inches='tight')
```

```python
n2=plt.figure(figsize=(1,1),dpi=200)
#large axes
axes1=n2.add_axes([0,0,1,1])
axes1.plot(a,b);
axes1.set_xlabel("A----->")
axes1.set_ylabel("B----->")
axes1.set_title("FIGURE 1")
#small axes
axes2=n2.add_axes([1,1,2,2])
axes2.plot(x,y);
axes2.set_xlabel("X----->")
axes2.set_ylabel("Y----->")
axes2.set_title("FIGURE 2");
n2.savefig('Figure 2',bbox_inches='tight')
```
