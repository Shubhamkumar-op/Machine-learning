```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,100,10)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2);
ax.plot(x,x**3);
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label="x**2")
ax.plot(x,x**3,label="x**3")
ax.legend();
fig.savefig('Naming the 2 figure',bbox_inches='tight')
```
