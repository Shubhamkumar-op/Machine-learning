```python
df
```

```python
import numpy as np
df.max()
```

```python
df['rt_nom']=np.round(df['RottenTomatoes']/20,1)
df['rt_nomu']=np.round(df['RottenTomatoes_User']/20,1)
df['met_nom']=np.round(df['Metacritic']/20,1)
df['met_nomu']=np.round(df['Metacritic_User']/2,1)
df['imbd_nom']=np.round(df['IMDB']/2,1)
```
