```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df=pd.read_csv('dm_office_sales.csv')
```

```python
df
```

```python
plt.figure(figsize=(16,8),dpi=200)
sns.scatterplot(x='sales',y='salary',data=df,hue='division');
plt.savefig('scatterplot1.png')
```

```python
plt.figure(figsize=(16,4),dpi=200)
sns.scatterplot(x='sales',y='salary',data=df,s=100,alpha=0.2);
plt.savefig('scatterplot2.png')
```

```python
plt.figure(figsize=(16,8),dpi=200)
sns.scatterplot(x='sales',y='salary',data=df,s=100,style='level of education',hue='level of education');
plt.savefig('scatterplot3.png')
```
