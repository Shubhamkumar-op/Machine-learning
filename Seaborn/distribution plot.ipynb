```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df=pd.read_csv('dm_office_sales.csv')
```

```python
df
```

```python
sns.rugplot(x='salary',data=df,height=0.5);
```

```python
sns.displot(data=df,x='salary',bins=40,kde=True)
```

```python
sns.kdeplot(data=df,x='salary',shade=True,)
```

```python
df
```

```python
df['division'].value_counts()
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(data=df,x='division');
```

```python
plt.figure(figsize=(14,6))
sns.barplot(data=df,x='level of education',y='salary',estimator=np.sum,hue='division')
plt.legend(bbox_to_anchor=(1.05,1))
```

```python
df=pd.read_csv('StudentsPerformance.csv')
```

```python
df
```

```python
plt.figure(figsize=(16,4))
sns.boxplot(data=df,y='math score',x='gender',hue='parental level of education')
plt.legend(bbox_to_anchor=(1.05,1));
```

```python
plt.figure(figsize=(16,8))
sns.boxenplot(y='math score',hue='parental level of education',data=df,x='gender')
plt.legend(bbox_to_anchor=(0.995,0.25));
```
